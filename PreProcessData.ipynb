```python
#The dataset that I have right now is quite vast and unnecessary for our goals of predicting stock outcomes.
#I'm going to focus more on the year 2023-2024 for preprocessing
#The idea is to let go of older data. The market has changed significantly, and I want our model to reflect to
#more relevant years
import pandas as pd
df = pd.read_csv("Walmart Retail_RAW.csv")
df["Date"] = pd.to_datetime(df['Date'])

#This is how I will be trimming down to the most recent year that the data provides, which is 2023-2024
filtered_df = df.loc[
    (df['Date'] >= '2023-01-01') & (df['Date'] <= '2024-01-02')]

#Reset the index with my new rows
filtered_df = filtered_df.reset_index(drop=True)

#Write the trimmed version to a new file
filtered_df.to_csv("retail_inventory_2023_2024.csv", index=False)
```
